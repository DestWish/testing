```python
%%writefile is_even.py
def is_even(number):
    return number % 2 == 0
```

```python
%%writefile test_is_even.py
import pytest
from is_even import is_even

@pytest.mark.parametrize("number, expected", [
    (2, True),
    (3, False),
    (0, True),
    (-2, True),
    (-3, False),
    (0.5, False),
    (4.0, True),
    (-4.0, True),
    (-0.5, False)])
def test_is_even(number, expected):
    assert is_even(number) == expected
```

```python
!pytest test_is_even.py -v
```

```python
%%writefile calculate_area.py
def calculate_area(length, width):
    return length * width
```

```python
%%writefile test_calculate_area.py
import pytest
from calculate_area import calculate_area
@pytest.mark.parametrize("length, width, expected", [
    (2, 3, 6),
    (0, 5, 0),
    (4, 4, 16),
    (7, 2, 14),
    (5, 0, 0),
    (3,-3, 9),
    (0.5, 10, 5.0)])
def test_calculate_area(length, width, expected):
    assert calculate_area(length, width) == expected
```

```python
!pytest test_calculate_area.py -v
```

```python
%%writefile classify_triangle.py
def classify_triangle(a, b, c):
    if a == b == c:
        return "Равносторонний"
    elif a == b or b == c or a == c:
        return "Равнобедренный"
    else:
        return "Разносторонний"
```

```python
%%writefile test_classify_triangle.py
import pytest
from classify_triangle import classify_triangle

@pytest.mark.parametrize("a, b, c, expected", [
    (3, 3, 3, "Равносторонний"),
    (4, 4, 4, "Равносторонний"),
    (4, 3, 3, "Равнобедренный"),
    (3, 3, 4, "Равнобедренный"),
    (3, 4, 5, "Разносторонний"),
    (0,0,1, "не треугольник"),
    (1,1,3, "не треугольник"),
    (2,2,4, "не треугольник")])
def test_classify_triangle(a, b, c, expected):
    assert classify_triangle(a, b, c) == expected
```

```python
!pytest test_classify_triangle.py -v
```
