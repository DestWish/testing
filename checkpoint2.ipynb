Метод принимающий строку и возвращающий ее длину. Тесты на пустую строку, многострочные, из пробелов.

```python
%%writefile strlen.py
def strlen(a: str) -> int:
    return len(a)
```

```python
%%writefile test_strlen.py
import pytest
from strlen import strlen

@pytest.mark.parametrize("a, expected", [
    ("", 0),
    ("     ", 5),
    ("1234567890", 10),
    ("1234567890" 
     "1234567890", 20),
])

def test_strlen(a, expected):
    assert strlen(a) == expected
```

```python
!pytest test_strlen.py -v
```

Метод сохраняющий строку в файл. 

```python
%%writefile strToNewFile.py
def strToNewFile(a: str, filename: str) -> None:
    with open(filename, 'w') as f:
        f.write(a)
```

```python
%%writefile test_strToNewFile.py
import pytest
from strToNewFile import strToNewFile




@pytest.mark.parametrize("a", [
    (""),
    ("     "),
    ("1234567890"),
    ("1234567890" 
     "1234567890"),
])
def test_strToNewFile(a):
    strToNewFile(a, "example.txt")
    with open("example.txt", "r", encoding="utf-8") as f:
        assert f.read() == a
```

```python
!pytest test_strToNewFile.py -v
```

```python
%%file test_sqlite.py
import pytest
import sqlite3

@pytest.fixture(scope="module")
def db():
    conn = sqlite3.connect(":memory:")
    cursor = conn.cursor()

    cursor.execute("""
        CREATE TABLE users (
            id INTEGER PRIMARY KEY,
            name TEXT
        )
    """)
    conn.commit()

    yield cursor

    cursor.execute("DROP TABLE users")
    conn.commit()
    conn.close()


def test_insert_user():
    db.execute("INSERT INTO users (name) VALUES (?)", ("Alice",))
    db.connection.commit()

    db.execute("SELECT name FROM users WHERE name = ?", ("Alice",))
    user = db.fetchone()
    assert user[0] == "Alice"
```
